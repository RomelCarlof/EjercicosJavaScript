```python
def mergeSort(alist):
    print("Splitting:", alist)

    if len(alist) > 1:
        mid = len(alist) // 2
        lefthalf = alist[:mid]
        righthalf = alist[mid:]

        mergeSort(lefthalf)
        mergeSort(righthalf)

        i = 0  # índice para lefthalf
        j = 0  # índice para righthalf
        k = 0  # índice para alist

        print(f"Comenzando fusión de {lefthalf} y {righthalf}")

        while i < len(lefthalf) and j < len(righthalf):
            print(f"Comparando {lefthalf[i]} y {righthalf[j]} -> ", end="")
            if lefthalf[i] < righthalf[j]:
                alist[k] = lefthalf[i]
                print(f"{lefthalf[i]} es menor → alist[{k}] = {lefthalf[i]}")
                i += 1
            else:
                alist[k] = righthalf[j]
                print(f"{righthalf[j]} es menor → alist[{k}] = {righthalf[j]}")
                j += 1
            k += 1
            print(f"i = {i}, j = {j}, k = {k}, alist = {alist}")

        while i < len(lefthalf):
            alist[k] = lefthalf[i]
            print(f"Copiando restante de izquierda: alist[{k}] = {lefthalf[i]}")
            i += 1
            k += 1
            print(f"i = {i}, k = {k}, alist = {alist}")

        while j < len(righthalf):
            alist[k] = righthalf[j]
            print(f"Copiando restante de derecha: alist[{k}] = {righthalf[j]}")
            j += 1
            k += 1
            print(f"j = {j}, k = {k}, alist = {alist}")

        print("Merging:", alist)
        print("-------")

# Lista a ordenar
alist = [54, 26, 93, 17, 77, 31, 44, 55, 20]
mergeSort(alist)
print("Lista final ordenada:", alist)
```
